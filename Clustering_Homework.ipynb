```python
import numpy as n
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv(r"C:\Users\Nare\Downloads\Mall_Customers.csv")
```

```python
data
```

```python
data['Gender'] = (data['Gender'] == 'Male' ).astype('int8')
data
```

```python
data = data.drop('CustomerID', axis = 1)
data
```

```python
ob = MinMaxScaler()
data = pd.DataFrame(ob.fit_transform(data), columns = data.columns, index = data.index)
```

```python
data
```

```python
X=data
```

# KMeans

```python
kmeans = KMeans(n_clusters=7, random_state=0, n_init="auto")
y_pred = kmeans.fit_predict(X)
```

```python
centers = kmeans.cluster_centers_
```

```python
sklearn_score = silhouette_score(X, y_pred)
```

```python
sklearn_score
```

# AgglomerativeClustering


```python
model = AgglomerativeClustering(n_clusters=5, linkage='complete')
y_pred = model.fit_predict(X)
sklearn_score = silhouette_score(X, y_pred)
```

```python
sklearn_score
```

# GMM

```python
gmm_sk = GaussianMixture(n_components=4, random_state=42)
gmm_sk.fit(X)
y_pred = gmm_sk.predict(X)
sklearn_score = silhouette_score(X, y_pred)
sklearn_score
```

# DBSCAN

```python
model = DBSCAN(eps=0.73, min_samples=3)
y_pred = model.fit_predict(X)
sklearn_score = silhouette_score(X, y_pred)
sklearn_score
```
